```python
listt = [
    [-5,  6],   # top-left
    [ 5,  6],   # top-right
    [ 5,  4],   # bar bottom-right
    [ 1,  4],   # stem top-right
    [ 1, -6],   # stem bottom-right
    [-1, -6],   # stem bottom-left
    [-1,  4],   # stem top-left
    [-5,  4]    # bar bottom-left
]
complex_shape = Geometry.polygon(*listt)
complex_shape.show()
```

```python
listt = [
    [0, 0],    # bottom-left
    [0, 10],   # top-left
    [3, 10],   # top of left stroke
    [7, 3],    # diagonal intersection
    [7, 10],   # top of right stroke
    [10, 10],  # top-right
    [10, 0],   # bottom-right
    [7, 0],    # bottom of right stroke
    [3, 7],    # diagonal intersection
    [3, 0],    # bottom of left stroke
]

complex_shape = Geometry.polygon(*listt)
print(complex_shape)
complex_shape.show()
```
